```python
import pandas as pd
import numpy as np
```

# Segmentacion de Data

```python
#Clientes con venta cuya coordenada hace match
df_cross_venta_match = df_cross[df_cross['flag_cross'] == 1]
df_cross_venta_match
```

```python
#Clientes con venta cuya coordenada no hace match
df_cross_venta_not_match = df_cross[(df_cross['flag_cross'] == 0) & (~df_cross.isnull().any(axis=1))]
df_cross_venta_not_match
```

```python
#Clientes con venta que presenta nulos
df_cross_venta_nulls = df_cross_nulls[(df_cross_nulls['departamen'].isnull()) & (df_cross_nulls['provincia'].isnull())
                                     & (df_cross_nulls['distrito'].isnull()) & (df_cross_nulls['name'].notna())]

df_cross_venta_nulls
```

```python
#Cliente sin venta con coordenadas
df_cross_not_venta = df_cross_nulls[(df_cross_nulls['departamen'].notna()) & (df_cross_nulls['provincia'].notna())
                                   & (df_cross_nulls['distrito'].notna()) & (df_cross_nulls['name'].isnull())]

df_cross_not_venta
```

```python
#Cliente sin venta con nulos
df_cross_not_venta_nulls = df_cross_nulls[(df_cross_nulls['departamen'].isnull()) & (df_cross_nulls['provincia'].isnull())
                                   & (df_cross_nulls['distrito'].isnull()) & (df_cross_nulls['name'].isnull())]

df_cross_not_venta_nulls
```

```python
print('Clientes con venta cuya coordenada hace match', df_cross_venta_match.shape[0])
print('Clientes con venta cuya coordenada no hace match', df_cross_venta_not_match.shape[0])
print('Clientes con venta que presenta nulos',df_cross_venta_nulls.shape[0])
print('Cliente sin venta con coordenadas validas(dentro del territorio peruano)',df_cross_not_venta.shape[0])
print('Cliente sin venta con nulos',df_cross_not_venta_nulls.shape[0])

print('\nTotal de Datos',df_cross.shape[0])
```

# Tratamiento clientes con venta que presentan nulos

```python
df_cross_venta_nulls
```

```python
df_cross_venta_null_dif_0 = df_cross_venta_nulls[(df_cross_venta_nulls['x'] != 0) & (df_cross_venta_nulls['y'] != 0)]
df_cross_venta_null_igual_0 = df_cross_venta_nulls[(df_cross_venta_nulls['x'] == 0) & (df_cross_venta_nulls['y'] == 0)]
df_cross_venta_null_X_dif_0 = df_cross_venta_nulls[(df_cross_venta_nulls['x'] != 0) & (df_cross_venta_nulls['y'] == 0)]
df_cross_venta_null_y_dif_0 = df_cross_venta_nulls[(df_cross_venta_nulls['x'] == 0) & (df_cross_venta_nulls['y'] != 0)]
```

```python
df_cross_venta_null_dif_0
```

```python
df_cross_venta_null_igual_0
```

```python
df_cross_venta_null_X_dif_0
```

```python
df_cross_venta_null_y_dif_0
```

```python
#Ambas coordenadas mayores a 0
temp1 = df_cross_venta_null_dif_0[(df_cross_venta_null_dif_0['x'] > 0) & (df_cross_venta_null_dif_0['y'] > 0)]

#Ambas coordenadas menores a 0
temp2 = df_cross_venta_null_dif_0[(df_cross_venta_null_dif_0['x'] < 0) & (df_cross_venta_null_dif_0['y'] < 0)]

# X mayor a 0, Y menor a 0
temp3 = df_cross_venta_null_dif_0[(df_cross_venta_null_dif_0['x'] > 0) & (df_cross_venta_null_dif_0['y'] < 0)]

# X menor a 0, Y mayor a 0
temp4 = df_cross_venta_null_dif_0[(df_cross_venta_null_dif_0['x'] < 0) & (df_cross_venta_null_dif_0['y'] > 0)]
```

```python
temp1
```

```python
for index, row in temp1.iterrows():
    if (row['x'] < row['y']):
        temp = row['x']
        temp1.at[index, 'x'] = row['y']
        temp1.at[index, 'y'] = temp
        
temp1['x'] = temp1['x'] * -1
temp1['y'] = temp1['y'] * -1

temp1
```

```python
temp2
```

```python
for index, row in temp2.iterrows():
    if (row['x'] > row['y']):
        temp = row['x']
        temp2.at[index, 'x'] = row['y']
        temp2.at[index, 'y'] = temp
        
temp2
```

```python
temp3
```

```python
temp3['x'] = temp3['x'] * -1

for index, row in temp3.iterrows():
    if (row['x'] > row['y']):
        temp = row['x']
        temp3.at[index, 'x'] = row['y']
        temp3.at[index, 'y'] = temp
        
temp3
```

```python
temp4
```

```python
temp4['y'] = temp4['y'] * -1

for index, row in temp4.iterrows():
    if (row['x'] > row['y']):
        temp = row['x']
        temp4.at[index, 'x'] = row['y']
        temp4.at[index, 'y'] = temp
        
temp4
```

```python
df_cross_venta_null_dif_0 = pd.concat([temp1,temp2,temp3,temp4])
df_cross_venta_null_dif_0
```
